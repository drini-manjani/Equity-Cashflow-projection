```python
"""
NAV Projection — “Institutional-grade” version (FIXED NaN propagation)

Key improvements vs your original:
1) Distributed-lag public market valuation model:
      omega_t = alpha + b0 * r_t + b1 * r_{t-1} + eps
   (captures NAV marking lag / appraisal smoothing)

2) Partial pooling / backoff ladder to avoid sparse (Strategy, Grade, AgeBucket):
   - Betas estimated at (Strategy, Grade) when enough data, else Strategy, else global.
   - Alpha uses (Strategy, Grade, AgeBucket) only when enough data, else (Strategy, Grade),
     else Strategy, else global.
   - Sigma estimated pooled and shrunk to avoid noisy small-sample vol.

3) Robust NaN handling:
   - Fix msci_ret_q_lag1 for FIRST projected quarter using last historical MSCI return.
   - Ensure msci_ret_q_lag1 has NO NaNs in projection horizon.
   - Ensure NAV_start is ALWAYS numeric (no NaNs), and fix NaN “or 0.0” bug.

Outputs:
- nav_projection_sota_*.csv and .parquet
- calibration diagnostics tables (betas, alphas, sigmas)

"""

import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# -----------------------------
# Utilities
# -----------------------------
AGE_LABELS = ["0-2y", "2-4y", "4-6y", "6-8y", "8-10y", "10-15y", "15-20y", "20y+"]
AGE_BINS_Q = [-1, 7, 15, 23, 31, 39, 59, 79, 10_000]  # quarters


def quarter_end_from_year_quarter(year: int, quarter: str) -> pd.Timestamp:
    q = quarter.upper().strip()
    if q not in {"Q1", "Q2", "Q3", "Q4"}:
        raise ValueError("Quarter must be one of: Q1, Q2, Q3, Q4")
    q_num = int(q[1])
    return pd.Period(f"{year}Q{q_num}", freq="Q").to_timestamp("Q")


def find_msci_projection_file(data_dir: str) -> str:
    cands = glob.glob(os.path.join(data_dir, "msci_projection_*.parquet")) + \
            glob.glob(os.path.join(data_dir, "msci_projection_*.csv"))
    if not cands:
        raise FileNotFoundError(
            f"No MSCI projection file found in {data_dir}. "
            "Expected msci_projection_*.parquet or msci_projection_*.csv"
        )
    cands.sort(key=os.path.getmtime, reverse=True)
    return cands[0]


def load_msci_projection(path: str) -> pd.DataFrame:
    if path.lower().endswith(".parquet"):
        df = pd.read_parquet(path)
    else:
        df = pd.read_csv(path)

    if not {"quarter_end", "msci_ret_q"}.issubset(df.columns):
        raise ValueError("MSCI projection file must contain columns: quarter_end, msci_ret_q")

    df = df.copy()
    df["quarter_end"] = pd.to_datetime(df["quarter_end"])
    df = df.sort_values("quarter_end").reset_index(drop=True)
    return df[["quarter_end", "msci_ret_q"]]


def add_quarters(qe: pd.Timestamp, q: float) -> pd.Timestamp:
    if pd.isna(qe):
        return pd.NaT
    p = pd.Period(qe, freq="Q")
    return (p + int(round(q))).to_timestamp("Q")


def make_age_bucket_q(age_q: pd.Series) -> pd.Categorical:
    return pd.cut(age_q, bins=AGE_BINS_Q, labels=AGE_LABELS)
```

```python
# -----------------------------
# Config — thresholds for “enough data”
# -----------------------------
NAV_COL = "NAV Adjusted EUR"
DRAW_COL = "Adj Drawdown EUR"
REPAY_COL = "Adj Repayment EUR"
SIZE_COL = "Target Fund Size"

# omega construction
NAV_EPS = 100.0       # nav_prev below this => unstable omega
OMEGA_CLIP = 0.8

# calibration minimums
MIN_FUNDS_BETA = 10
MIN_OBS_BETA = 80

MIN_FUNDS_ALPHA_BUCKET = 6
MIN_OBS_ALPHA_BUCKET = 60

# volatility shrinkage strength (bigger => more shrinkage to pooled)
SIGMA_SHRINK_K = 120.0

# NAV-start imputation
DRAW_EPS = 1000.0
SIZE_EPS = 1e6
MIN_OBS_RATIO = 50
```

```python
# -----------------------------
# Main inputs / paths
# -----------------------------
year = int(input("Enter year (e.g. 2025): ").strip())
quarter = input("Enter quarter (Q1, Q2, Q3, Q4): ").strip().upper()

n_q = int(input("Enter number of quarters to project (0 => default 40): ").strip() or "0")
if n_q == 0:
    n_q = 40

BASE_DIR = os.path.join("C:\\Users", os.environ.get("USERNAME"), "Documents", "Equity")
HOME = os.path.join(BASE_DIR, f"{year}_{quarter}")
DATA_DIR = os.path.join(HOME, "data")

data_path = os.path.join(DATA_DIR, "data.parquet")
if not os.path.exists(data_path):
    raise FileNotFoundError(f"Missing data.parquet at: {data_path}")

msci_xlsx = os.path.join(DATA_DIR, "MSCI.xlsx")
if not os.path.exists(msci_xlsx):
    raise FileNotFoundError(f"Missing MSCI.xlsx at: {msci_xlsx}")

msci_proj_path = find_msci_projection_file(DATA_DIR)
print("Using MSCI projection file:", msci_proj_path)

start_qe = quarter_end_from_year_quarter(year, quarter)
print("Projection start quarter-end:", start_qe.date())
```

```python
# -----------------------------
# Load fund panel
# -----------------------------
data = pd.read_parquet(data_path).copy()

data["quarter_end"] = pd.PeriodIndex(
    data["Year of Transaction Date"].astype(int).astype(str) + "Q" +
    data["Quarter of Transaction Date"].astype(int).astype(str),
    freq="Q"
).to_timestamp("Q")
data["quarter_end"] = pd.to_datetime(data["quarter_end"])

required_cols = [
    "FundID",
    "quarter_end",
    NAV_COL,
    DRAW_COL,
    REPAY_COL,
    "Adj Strategy",
    "Grade",
    SIZE_COL,
    "Fund_Age_Quarters",
    "Planned end date with add. years as per legal doc",
]
missing = [c for c in required_cols if c not in data.columns]
if missing:
    raise ValueError(f"Missing required columns in data.parquet: {missing}")

data["planned_end_qe"] = pd.to_datetime(
    data["Planned end date with add. years as per legal doc"],
    errors="coerce"
).dt.to_period("Q").dt.to_timestamp("Q")

data[NAV_COL] = pd.to_numeric(data[NAV_COL], errors="coerce")
for c in [DRAW_COL, REPAY_COL, SIZE_COL, "Fund_Age_Quarters"]:
    data[c] = pd.to_numeric(data[c], errors="coerce").fillna(0.0)

data = data.sort_values(["FundID", "quarter_end"]).reset_index(drop=True)

print("Loaded data rows:", len(data))
print("Funds:", data["FundID"].nunique())
print("Missing NAV rows:", data[NAV_COL].isna().sum())
print("Zero NAV rows:", (data[NAV_COL] == 0).sum())
```

```python
# -----------------------------
# MSCI history: quarterly returns + lag
# -----------------------------
msci_raw = pd.read_excel(msci_xlsx)[["Date", "SCXP Index"]].copy()
msci_raw["Date"] = pd.to_datetime(msci_raw["Date"], errors="coerce")
msci_raw["SCXP Index"] = pd.to_numeric(msci_raw["SCXP Index"], errors="coerce")
msci_raw = msci_raw.dropna().sort_values("Date")

msci_raw["quarter_end"] = msci_raw["Date"].dt.to_period("Q").dt.to_timestamp("Q")
msci_hist = (
    msci_raw.groupby("quarter_end", as_index=False)["SCXP Index"]
    .last()
    .rename(columns={"SCXP Index": "index_level"})
)
msci_hist["msci_ret_q"] = msci_hist["index_level"].pct_change()
msci_hist = msci_hist.dropna(subset=["msci_ret_q"])[["quarter_end", "msci_ret_q"]]
msci_hist["quarter_end"] = pd.to_datetime(msci_hist["quarter_end"])
msci_hist = msci_hist.sort_values("quarter_end").reset_index(drop=True)

msci_hist["msci_ret_q_lag1"] = msci_hist["msci_ret_q"].shift(1)

print("MSCI history quarters:", len(msci_hist))
```

```python
# -----------------------------
# MSCI future projection (+ lag) and CRITICAL lag fix for first projected quarter
# -----------------------------
msci_proj = load_msci_projection(msci_proj_path)
msci_proj = msci_proj.sort_values("quarter_end").reset_index(drop=True)
msci_proj["msci_ret_q_lag1"] = msci_proj["msci_ret_q"].shift(1)

msci_future = msci_proj.loc[msci_proj["quarter_end"] > start_qe].head(n_q).copy()
if msci_future.empty:
    raise ValueError("MSCI projection contains no quarters after the start date you selected.")

# ---- FIX: lag for first projected quarter from last historical realized return (<= start_qe) ----
last_hist_ret = msci_hist.loc[msci_hist["quarter_end"] <= start_qe, "msci_ret_q"].tail(1)
if len(last_hist_ret):
    msci_future.loc[msci_future.index[0], "msci_ret_q_lag1"] = float(last_hist_ret.iloc[0])
else:
    msci_future.loc[msci_future.index[0], "msci_ret_q_lag1"] = 0.0

# ensure no NaNs in lag column
msci_future["msci_ret_q_lag1"] = msci_future["msci_ret_q_lag1"].fillna(0.0)

print("Future MSCI quarters available:", len(msci_future))
print(msci_future.head())
```

```python
# -----------------------------
# Planned end overrun by strategy
# -----------------------------
last_obs = data.groupby("FundID")["quarter_end"].max().rename("last_qe")
fund_static = data.sort_values(["FundID", "quarter_end"]).groupby("FundID").tail(1).copy()
fund_static = fund_static.merge(last_obs, on="FundID", how="left")

def quarters_diff(a: pd.Timestamp, b: pd.Timestamp) -> float:
    if pd.isna(a) or pd.isna(b):
        return np.nan
    return float(pd.Period(a, freq="Q").ordinal - pd.Period(b, freq="Q").ordinal)

fund_static["overrun_q"] = fund_static.apply(
    lambda r: max(quarters_diff(r["last_qe"], r["planned_end_qe"]), 0.0)
    if pd.notna(r["planned_end_qe"]) else np.nan,
    axis=1
)

fund_static["ever_overran"] = fund_static["overrun_q"].fillna(0) > 0
ever_overran_map = fund_static.set_index("FundID")["ever_overran"]

overran_only = fund_static.loc[fund_static["overrun_q"].notna() & (fund_static["overrun_q"] > 0)].copy()
avg_overrun_by_strategy = overran_only.groupby("Adj Strategy")["overrun_q"].mean().clip(lower=0.0)
avg_overrun_by_strategy = avg_overrun_by_strategy.reindex(
    fund_static["Adj Strategy"].dropna().unique(),
    fill_value=0.0
)

print("Strategies with observed overruns:", overran_only["Adj Strategy"].nunique())
print("Average overrun quarters by Adj Strategy (head):")
print(avg_overrun_by_strategy.head(10))
```

```python
# -----------------------------
# Build omega (valuation return net of flows)
# -----------------------------
df = data.copy().sort_values(["FundID", "quarter_end"]).reset_index(drop=True)

df["nav_prev"] = df.groupby("FundID")[NAV_COL].shift(1)
df["flow_net"] = df[DRAW_COL] - df[REPAY_COL]

m = df["nav_prev"].abs() > NAV_EPS
df["omega"] = np.nan
df.loc[m, "omega"] = ((df.loc[m, NAV_COL] - df.loc[m, "nav_prev"]) - df.loc[m, "flow_net"]) / df.loc[m, "nav_prev"]
df["omega"] = df["omega"].clip(lower=-OMEGA_CLIP, upper=OMEGA_CLIP)

print("Omega available observations:", df["omega"].notna().sum())
```

```python
# -----------------------------
# Calibration dataset (omega + MSCI with lag)
# -----------------------------
cal = df.merge(msci_hist, on="quarter_end", how="left")
cal = cal.dropna(subset=["omega", "msci_ret_q", "msci_ret_q_lag1"]).copy()
cal["AgeBucket"] = make_age_bucket_q(cal["Fund_Age_Quarters"])

cal = cal[[
    "FundID", "Adj Strategy", "Grade", "AgeBucket",
    "omega", "msci_ret_q", "msci_ret_q_lag1"
]].copy()

print("Calibration rows (omega & msci+lag avail):", len(cal))
```

```python
import numpy as np
import pandas as pd
from math import sqrt
from scipy import stats  # pip-installed in most envs; if missing, tell me and I'll remove dependency.

# ============================================================
# Cluster-robust OLS (FundID clusters)
# ============================================================
def ols_cluster_robust(df, y_col, x_cols, cluster_col):
    """
    Fits OLS y = a + Xb and returns cluster-robust SEs clustered by cluster_col.
    Uses df = (#clusters - 1) for t-tests (common in practice).

    Returns dict with:
      coef (Series), se (Series), t (Series), p (Series),
      n_obs, n_clusters, df_dof
    """
    d = df.dropna(subset=[y_col] + x_cols + [cluster_col]).copy()
    n = len(d)
    if n == 0:
        return None

    # Build design matrix with intercept
    y = d[y_col].to_numpy(float)
    X = np.column_stack([np.ones(n)] + [d[c].to_numpy(float) for c in x_cols])
    k = X.shape[1]

    # OLS coefficients
    XtX = X.T @ X
    XtX_inv = np.linalg.pinv(XtX)
    beta = XtX_inv @ (X.T @ y)

    # Residuals
    u = y - X @ beta

    # Clusters
    clusters = d[cluster_col].to_numpy()
    uniq = pd.unique(clusters)
    G = len(uniq)
    df_dof = G - 1

    # If too few clusters, clustered inference not defined -> return None
    if G <= 1:
        return None

    # Meat: sum over clusters of (X_g' u_g)(X_g' u_g)'
    meat = np.zeros((k, k), dtype=float)
    for g in uniq:
        mask = (clusters == g)
        Xg = X[mask, :]
        ug = u[mask]
        Xgu = Xg.T @ ug  # k-vector
        meat += np.outer(Xgu, Xgu)

    # Cluster-robust covariance
    V = XtX_inv @ meat @ XtX_inv

    # Finite sample correction (common CR1)
    # scale = (G/(G-1)) * ((n-1)/(n-k))
    scale = (G / (G - 1)) * ((n - 1) / max(n - k, 1))
    V *= scale

    se = np.sqrt(np.diag(V))
    tstats = beta / se

    # Two-sided p-values using t(df=G-1)
    pvals = 2.0 * (1.0 - stats.t.cdf(np.abs(tstats), df=df_dof))

    names = ["alpha"] + x_cols
    out = {
        "coef": pd.Series(beta, index=names),
        "se": pd.Series(se, index=names),
        "t": pd.Series(tstats, index=names),
        "p": pd.Series(pvals, index=names),
        "n_obs": int(n),
        "n_clusters": int(G),
        "df_dof": int(df_dof),
    }
    return out


def cluster_mean_stats(df, y_col, cluster_col):
    """
    Mean with fund-cluster-robust SE via intercept-only regression:
      y = alpha + e
    """
    res = ols_cluster_robust(df, y_col=y_col, x_cols=[], cluster_col=cluster_col)
    if res is None:
        return None
    return {
        "alpha": float(res["coef"]["alpha"]),
        "se_alpha": float(res["se"]["alpha"]),
        "t_alpha": float(res["t"]["alpha"]),
        "p_alpha": float(res["p"]["alpha"]),
        "n_obs": res["n_obs"],
        "n_funds": res["n_clusters"],
        "df": res["df_dof"],
    }


# ============================================================
# Statistical decision settings (NOT “hardcoded thresholds”)
# ============================================================
alpha_level = float(input("Significance level for using group params? (e.g. 0.10 or 0.05) [0.10]: ").strip() or "0.10")

# Optional: require a minimum number of clusters to trust clustered t-tests.
# This is not arbitrary “model tuning”; it’s a validity condition.
min_clusters_for_inference = int(input("Minimum #fund clusters for clustered inference [8]: ").strip() or "8")


# ============================================================
# 1) Estimate betas with cluster-robust inference
#    omega = alpha + b0*r_t + b1*r_{t-1} + eps
# ============================================================
# Ensure lag exists in cal
if "msci_ret_q_lag1" not in cal.columns:
    raise ValueError("cal must include msci_ret_q_lag1")

beta_rows_sg = []
for (s, g), grp in cal.groupby(["Adj Strategy", "Grade"], dropna=False):
    res = ols_cluster_robust(
        grp, y_col="omega",
        x_cols=["msci_ret_q", "msci_ret_q_lag1"],
        cluster_col="FundID"
    )
    if res is None:
        continue

    n_funds = res["n_clusters"]
    # only decide “use” when clustered inference is meaningful
    usable = (n_funds >= min_clusters_for_inference)

    b0 = float(res["coef"]["msci_ret_q"])
    b1 = float(res["coef"]["msci_ret_q_lag1"])
    p0 = float(res["p"]["msci_ret_q"])
    p1 = float(res["p"]["msci_ret_q_lag1"])

    # Decision purely statistical: at least one macro coefficient significant
    use_beta = bool(usable and ((p0 < alpha_level) or (p1 < alpha_level)))

    beta_rows_sg.append({
        "Adj Strategy": s,
        "Grade": g,
        "b0": b0, "b1": b1,
        "se_b0": float(res["se"]["msci_ret_q"]),
        "se_b1": float(res["se"]["msci_ret_q_lag1"]),
        "t_b0": float(res["t"]["msci_ret_q"]),
        "t_b1": float(res["t"]["msci_ret_q_lag1"]),
        "p_b0": p0,
        "p_b1": p1,
        "n_obs": res["n_obs"],
        "n_funds": n_funds,
        "df": res["df_dof"],
        "use_beta": use_beta
    })

beta_sg = pd.DataFrame(beta_rows_sg)

# Strategy-only betas
beta_rows_s = []
for s, grp in cal.groupby(["Adj Strategy"], dropna=False):
    res = ols_cluster_robust(
        grp, y_col="omega",
        x_cols=["msci_ret_q", "msci_ret_q_lag1"],
        cluster_col="FundID"
    )
    if res is None:
        continue

    n_funds = res["n_clusters"]
    usable = (n_funds >= min_clusters_for_inference)

    b0 = float(res["coef"]["msci_ret_q"])
    b1 = float(res["coef"]["msci_ret_q_lag1"])
    p0 = float(res["p"]["msci_ret_q"])
    p1 = float(res["p"]["msci_ret_q_lag1"])

    use_beta = bool(usable and ((p0 < alpha_level) or (p1 < alpha_level)))

    beta_rows_s.append({
        "Adj Strategy": s,
        "b0": b0, "b1": b1,
        "se_b0": float(res["se"]["msci_ret_q"]),
        "se_b1": float(res["se"]["msci_ret_q_lag1"]),
        "t_b0": float(res["t"]["msci_ret_q"]),
        "t_b1": float(res["t"]["msci_ret_q_lag1"]),
        "p_b0": p0,
        "p_b1": p1,
        "n_obs": res["n_obs"],
        "n_funds": n_funds,
        "df": res["df_dof"],
        "use_beta": use_beta
    })

beta_s = pd.DataFrame(beta_rows_s)

# Global betas
res_g = ols_cluster_robust(
    cal, y_col="omega",
    x_cols=["msci_ret_q", "msci_ret_q_lag1"],
    cluster_col="FundID"
)
if res_g is None:
    raise ValueError("Global beta regression failed unexpectedly.")

b0_g = float(res_g["coef"]["msci_ret_q"])
b1_g = float(res_g["coef"]["msci_ret_q_lag1"])

print("\nBeta coverage (cluster-robust):")
print("  (Strategy,Grade) groups estimated:", len(beta_sg))
print("  (Strategy) groups estimated:", len(beta_s))
print("  Global betas:", (b0_g, b1_g))
print("  (Strategy,Grade) use_beta==True:", int(beta_sg["use_beta"].sum()) if len(beta_sg) else 0)
print("  (Strategy) use_beta==True:", int(beta_s["use_beta"].sum()) if len(beta_s) else 0)

# Build lookup maps filtered by use_beta
beta_sg_use = beta_sg.loc[beta_sg["use_beta"]].set_index(["Adj Strategy", "Grade"])[["b0", "b1"]].to_dict("index")
beta_s_use = beta_s.loc[beta_s["use_beta"]].set_index(["Adj Strategy"])[["b0", "b1"]].to_dict("index")

def get_betas(strategy, grade):
    k = (strategy, grade)
    if k in beta_sg_use:
        d = beta_sg_use[k]
        return float(d["b0"]), float(d["b1"]), "sg_sig"
    if strategy in beta_s_use:
        d = beta_s_use[strategy]
        return float(d["b0"]), float(d["b1"]), "s_sig"
    return float(b0_g), float(b1_g), "global"


# ============================================================
# 2) Alpha estimation: macro-adjusted omega, then clustered mean tests
# ============================================================
cal2 = cal.copy()

# compute omega_adj using statistically selected betas
b0_used = []
b1_used = []
src_used = []
for _, r in cal2.iterrows():
    b0, b1, src = get_betas(r["Adj Strategy"], r["Grade"])
    b0_used.append(b0); b1_used.append(b1); src_used.append(src)

cal2["b0_used"] = b0_used
cal2["b1_used"] = b1_used
cal2["beta_source"] = src_used

cal2["omega_adj"] = cal2["omega"] - cal2["b0_used"] * cal2["msci_ret_q"] - cal2["b1_used"] * cal2["msci_ret_q_lag1"]

# Alpha by (Strategy, Grade, AgeBucket): clustered mean + p-value
alpha_rows_sga = []
for (s, g, a), grp in cal2.groupby(["Adj Strategy", "Grade", "AgeBucket"], dropna=False):
    st = cluster_mean_stats(grp, y_col="omega_adj", cluster_col="FundID")
    if st is None:
        continue
    use_alpha = bool((st["n_funds"] >= min_clusters_for_inference) and (st["p_alpha"] < alpha_level))
    alpha_rows_sga.append({"Adj Strategy": s, "Grade": g, "AgeBucket": a, **st, "use_alpha": use_alpha})

alpha_sga = pd.DataFrame(alpha_rows_sga)

# Alpha by (Strategy, Grade)
alpha_rows_sg = []
for (s, g), grp in cal2.groupby(["Adj Strategy", "Grade"], dropna=False):
    st = cluster_mean_stats(grp, y_col="omega_adj", cluster_col="FundID")
    if st is None:
        continue
    use_alpha = bool((st["n_funds"] >= min_clusters_for_inference) and (st["p_alpha"] < alpha_level))
    alpha_rows_sg.append({"Adj Strategy": s, "Grade": g, **st, "use_alpha": use_alpha})
alpha_sg = pd.DataFrame(alpha_rows_sg)

# Alpha by Strategy
alpha_rows_s = []
for s, grp in cal2.groupby(["Adj Strategy"], dropna=False):
    st = cluster_mean_stats(grp, y_col="omega_adj", cluster_col="FundID")
    if st is None:
        continue
    use_alpha = bool((st["n_funds"] >= min_clusters_for_inference) and (st["p_alpha"] < alpha_level))
    alpha_rows_s.append({"Adj Strategy": s, **st, "use_alpha": use_alpha})
alpha_s = pd.DataFrame(alpha_rows_s)

# Global alpha
st_g = cluster_mean_stats(cal2, y_col="omega_adj", cluster_col="FundID")
alpha_global = float(st_g["alpha"]) if st_g else 0.0

print("\nAlpha coverage (cluster-robust):")
print("  (Strategy,Grade,AgeBucket) estimated:", len(alpha_sga))
print("  use_alpha==True:", int(alpha_sga["use_alpha"].sum()) if len(alpha_sga) else 0)
print("  Global alpha:", alpha_global)

# Build lookup maps filtered by use_alpha
alpha_sga_use = alpha_sga.loc[alpha_sga["use_alpha"]].set_index(["Adj Strategy", "Grade", "AgeBucket"])["alpha"].to_dict()
alpha_sg_use = alpha_sg.loc[alpha_sg["use_alpha"]].set_index(["Adj Strategy", "Grade"])["alpha"].to_dict()
alpha_s_use = alpha_s.loc[alpha_s["use_alpha"]].set_index(["Adj Strategy"])["alpha"].to_dict()

def get_alpha(strategy, grade, age_bucket):
    k = (strategy, grade, age_bucket)
    if k in alpha_sga_use:
        return float(alpha_sga_use[k]), "sga_sig"
    k2 = (strategy, grade)
    if k2 in alpha_sg_use:
        return float(alpha_sg_use[k2]), "sg_sig"
    if strategy in alpha_s_use:
        return float(alpha_s_use[strategy]), "s_sig"
    return float(alpha_global), "global"
```

```python
# -----------------------------
# Parameter estimation: OLS omega = alpha + b0*r + b1*r_lag + eps
# -----------------------------
def ols_params(x0: np.ndarray, x1: np.ndarray, y: np.ndarray):
    X = np.column_stack([np.ones(len(y)), x0, x1])
    coef, *_ = np.linalg.lstsq(X, y, rcond=None)
    a, b0, b1 = coef
    resid = y - X @ coef
    sigma = float(np.std(resid, ddof=1)) if len(y) > 3 else float(np.std(resid)) if len(y) > 1 else 0.10
    sigma = max(sigma, 0.02)
    return float(a), float(b0), float(b1), float(sigma)


def summarize_group(g: pd.DataFrame) -> dict:
    n_obs = len(g)
    n_funds = g["FundID"].nunique()
    vx = float(np.var(g["msci_ret_q"].to_numpy(), ddof=1)) if n_obs > 1 else 0.0
    return {"n_obs": n_obs, "n_funds": n_funds, "var_msci": vx}
```

```python
# -----------------------------
# NAV-start imputation (FIXED NaN handling)
# -----------------------------
hist_upto["draw_cum"] = hist_upto.groupby("FundID")[DRAW_COL].cumsum()

if "draw_cum" not in base_rows.columns:
    base_rows = base_rows.merge(
        hist_upto.groupby("FundID")["draw_cum"].last().reset_index(),
        on="FundID",
        how="left"
    )

tmp = hist_upto.copy()
tmp["AgeBucket"] = make_age_bucket_q(tmp["Fund_Age_Quarters"])

tmp["ratio_nav_draw"] = np.where(
    (tmp[NAV_COL].notna()) & (tmp[NAV_COL].abs() > NAV_EPS) & (tmp["draw_cum"] > DRAW_EPS),
    tmp[NAV_COL] / tmp["draw_cum"],
    np.nan
)
tmp["ratio_nav_size"] = np.where(
    (tmp[NAV_COL].notna()) & (tmp[NAV_COL].abs() > NAV_EPS) & (tmp[SIZE_COL] > SIZE_EPS),
    tmp[NAV_COL] / tmp[SIZE_COL],
    np.nan
)

tmp["log_ratio_nav_draw"] = np.log(tmp["ratio_nav_draw"])
tmp["log_ratio_nav_size"] = np.log(tmp["ratio_nav_size"])
tmp.loc[~np.isfinite(tmp["log_ratio_nav_draw"]), "log_ratio_nav_draw"] = np.nan
tmp.loc[~np.isfinite(tmp["log_ratio_nav_size"]), "log_ratio_nav_size"] = np.nan

ratio_key = ["Adj Strategy", "Grade", "AgeBucket"]

def fit_lognorm(df: pd.DataFrame, col: str) -> pd.Series:
    g = df[col].dropna()
    if len(g) < MIN_OBS_RATIO:
        return pd.Series({"mu": np.nan, "sig": np.nan, "n": len(g)})
    return pd.Series({"mu": float(g.mean()), "sig": float(g.std(ddof=1)), "n": len(g)})

ratio_draw = tmp.groupby(ratio_key, dropna=False).apply(lambda g: fit_lognorm(g, "log_ratio_nav_draw")).reset_index()
ratio_size = tmp.groupby(ratio_key, dropna=False).apply(lambda g: fit_lognorm(g, "log_ratio_nav_size")).reset_index()

gdraw = ratio_draw.dropna(subset=["mu", "sig"])
gsize = ratio_size.dropna(subset=["mu", "sig"])
fallback_draw = {"mu": float(gdraw["mu"].median()) if len(gdraw) else 0.0,
                 "sig": float(gdraw["sig"].median()) if len(gdraw) else 0.75}
fallback_size = {"mu": float(gsize["mu"].median()) if len(gsize) else -2.0,
                 "sig": float(gsize["sig"].median()) if len(gsize) else 0.75}

ratio_draw_map = ratio_draw.set_index(ratio_key)[["mu", "sig", "n"]].to_dict("index")
ratio_size_map = ratio_size.set_index(ratio_key)[["mu", "sig", "n"]].to_dict("index")

def lookup_ratio(map_, strategy, grade, age_bucket, fallback):
    k = (strategy, grade, age_bucket)
    if k in map_:
        d = map_[k]
        if pd.notna(d["mu"]) and pd.notna(d["sig"]) and d["n"] >= MIN_OBS_RATIO:
            return float(d["mu"]), float(d["sig"]), "bucket"
    return float(fallback["mu"]), float(fallback["sig"]), "global"

rng_init = np.random.default_rng(2025)
base_rows["NAV_start"] = base_rows[NAV_COL]
base_rows["NAV_start_source"] = "observed"

for idx, r in base_rows.iterrows():
    nav_obs = r["NAV_start"]
    if pd.notna(nav_obs) and abs(nav_obs) > NAV_EPS:
        continue

    draw_cum = r.get("draw_cum", 0.0)
    size = r.get(SIZE_COL, 0.0)
    draw_cum = 0.0 if pd.isna(draw_cum) else float(draw_cum)
    size = 0.0 if pd.isna(size) else float(size)

    strategy = r["Adj Strategy"]
    # start grade = last observed AssignedGrade if available, else use Grade
    grade = r.get("AssignedGrade", r["Grade"])
    if pd.isna(grade):
        grade = r["Grade"]
    age_bucket = r["AgeBucket"]

    if draw_cum > DRAW_EPS:
        mu, sig, src = lookup_ratio(ratio_draw_map, strategy, grade, age_bucket, fallback_draw)
        ratio = float(np.exp(mu + sig * rng_init.standard_normal()))
        ratio = float(np.clip(ratio, 0.05, 5.0))
        base_rows.at[idx, "NAV_start"] = ratio * draw_cum
        base_rows.at[idx, "NAV_start_source"] = f"imputed_draw_{src}"
    elif size > SIZE_EPS:
        mu, sig, src = lookup_ratio(ratio_size_map, strategy, grade, age_bucket, fallback_size)
        ratio = float(np.exp(mu + sig * rng_init.standard_normal()))
        ratio = float(np.clip(ratio, 0.01, 2.0))
        base_rows.at[idx, "NAV_start"] = ratio * size
        base_rows.at[idx, "NAV_start_source"] = f"imputed_size_{src}"
    else:
        base_rows.at[idx, "NAV_start"] = 0.0
        base_rows.at[idx, "NAV_start_source"] = "imputed_zero_noinfo"

# ---- FIX: guarantee NAV_start is numeric (no NaN contagion) ----
base_rows["NAV_start"] = pd.to_numeric(base_rows["NAV_start"], errors="coerce").fillna(0.0)

print("\nNAV_start_source counts:")
print(base_rows["NAV_start_source"].value_counts().head(20))
print("NAV_start NaNs:", base_rows["NAV_start"].isna().sum())
```

```python
# -----------------------------
# ------------------------------------------------------------
# Grade transition matrix helpers (quarterly grade updates)
# ------------------------------------------------------------
from pathlib import Path

GRADE_STATES = ["A", "B", "C", "D"]
GRADE_TO_IDX = {g:i for i,g in enumerate(GRADE_STATES)}
IDX_TO_GRADE = {i:g for g,i in GRADE_TO_IDX.items()}
LAPLACE = 1.0
MIN_TRANSITIONS = 30
TARGET_BLEND = 0.20  # blend to historical grade distribution
DEFAULT_YEARLY_CHANGE = 0.25  # fallback if insufficient data

def _row_normalize(mat):
    mat = np.where(np.isfinite(mat), mat, 0.0)
    mat = np.clip(mat, 0.0, None)
    rs = mat.sum(axis=1, keepdims=True)
    rs = np.where(rs == 0, 1.0, rs)
    return mat / rs

def _laplace_smooth(counts, alpha=1.0):
    return counts + alpha

def build_transition_from_quarters(fund_quarters, strategy=None):
    """
    Build 1-quarter transition counts + probs from fund_quarters AssignedGrade.
    Uses consecutive quarters within each fund (RankQ -> RankQ+1).
    If strategy is provided, only that strategy is used.
    Returns (counts_df, probs_df, n_transitions)
    """
    fq = fund_quarters.copy()
    strat_col = "AdjStrategy" if "AdjStrategy" in fq.columns else ("Adj Strategy" if "Adj Strategy" in fq.columns else None)
    if strategy is not None and strat_col is not None:
        fq = fq[fq[strat_col] == strategy].copy()

    fq = fq.sort_values(["FundID", "RankQ"])
    fq["NextGrade"] = fq.groupby("FundID")["AssignedGrade"].shift(-1)

    trans = fq.dropna(subset=["AssignedGrade", "NextGrade"])
    trans = trans[trans["AssignedGrade"].isin(GRADE_STATES) & trans["NextGrade"].isin(GRADE_STATES)]

    n_transitions = len(trans)

    counts = pd.crosstab(
        trans["AssignedGrade"], trans["NextGrade"]
    ).reindex(index=GRADE_STATES, columns=GRADE_STATES, fill_value=0)

    counts = counts.astype(float)
    counts_sm = _laplace_smooth(counts.values, alpha=LAPLACE)
    probs = _row_normalize(counts_sm)

    # Calibrate to historical grade distribution (reduces erratic jumps)
    if len(fq):
        target = fq["AssignedGrade"].value_counts().reindex(GRADE_STATES, fill_value=0).to_numpy(dtype=float)
        if target.sum() > 0:
            target = target / target.sum()
            probs = (1.0 - TARGET_BLEND) * probs + TARGET_BLEND * np.tile(target, (len(GRADE_STATES), 1))
            probs = _row_normalize(probs)

    counts_df = pd.DataFrame(counts, index=GRADE_STATES, columns=GRADE_STATES)
    probs_df = pd.DataFrame(probs, index=GRADE_STATES, columns=GRADE_STATES)

    return counts_df, probs_df, n_transitions

def build_yearly_change_prob(fund_quarters, strategy=None):
    """Average probability of grade change per year (4 quarters) by strategy."""
    fq = fund_quarters.copy()
    strat_col = "AdjStrategy" if "AdjStrategy" in fq.columns else ("Adj Strategy" if "Adj Strategy" in fq.columns else None)
    if strategy is not None and strat_col is not None:
        fq = fq[fq[strat_col] == strategy].copy()
    if fq.empty:
        return DEFAULT_YEARLY_CHANGE

    probs = []
    for _, g in fq.sort_values(["FundID", "RankQ"]).groupby("FundID"):
        grades = g["AssignedGrade"].dropna().tolist()
        if len(grades) < 5:
            continue
        yearly = grades[::4]
        if len(yearly) < 2:
            continue
        changes = sum(1 for i in range(1, len(yearly)) if yearly[i] != yearly[i-1])
        probs.append(changes / (len(yearly) - 1))
    if not probs:
        return DEFAULT_YEARLY_CHANGE
    return float(np.mean(probs))

# --- Build matrices: strategy-specific (PE, VC) + fallback (All) ---
# fund_quarters must exist (from transition_matrix logic)
if "fund_quarters" not in locals():
    # Try to load from DATA_DIR if available
    _fq_path_parquet = os.path.join(DATA_DIR, "fund_quarters.parquet") if "DATA_DIR" in locals() else None
    _fq_path_csv = os.path.join(DATA_DIR, "fund_quarters.csv") if "DATA_DIR" in locals() else None
    if _fq_path_parquet and os.path.exists(_fq_path_parquet):
        fund_quarters = pd.read_parquet(_fq_path_parquet)
    elif _fq_path_csv and os.path.exists(_fq_path_csv):
        fund_quarters = pd.read_csv(_fq_path_csv)
    else:
        # Fallback: search BASE_DIR for the most recent fund_quarters file
        if "BASE_DIR" in locals() and os.path.exists(BASE_DIR):
            _candidates = []
            for root, dirs, files in os.walk(BASE_DIR):
                if "fund_quarters.parquet" in files:
                    _candidates.append(os.path.join(root, "fund_quarters.parquet"))
                elif "fund_quarters.csv" in files:
                    _candidates.append(os.path.join(root, "fund_quarters.csv"))
            if _candidates:
                _candidates.sort(key=lambda p: os.path.getmtime(p), reverse=True)
                _path = _candidates[0]
                print(f"Using fund_quarters from: {_path}")
                fund_quarters = pd.read_parquet(_path) if _path.endswith(".parquet") else pd.read_csv(_path)
        if "fund_quarters" not in locals():
            print("DATA_DIR:", DATA_DIR if "DATA_DIR" in locals() else None)
            print("BASE_DIR:", BASE_DIR if "BASE_DIR" in locals() else None)
            if "DATA_DIR" in locals():
                _p = os.path.join(DATA_DIR, "fund_quarters.parquet")
                _c = os.path.join(DATA_DIR, "fund_quarters.csv")
                print("Exists parquet:", os.path.exists(_p))
                print("Exists csv:", os.path.exists(_c))
            raise ValueError(
                "fund_quarters is required for transition matrices (AssignedGrade). "
                "Run transition_matrix.ipynb and save fund_quarters to DATA_DIR as fund_quarters.parquet or .csv."
            )

# All-portfolio fallback
all_counts, all_probs, all_n = build_transition_from_quarters(fund_quarters, strategy=None)

# Strategy-specific
pe_counts, pe_probs, pe_n = build_transition_from_quarters(fund_quarters, strategy="Private Equity")
vc_counts, vc_probs, vc_n = build_transition_from_quarters(fund_quarters, strategy="Venture Capital")
pe_pchg = build_yearly_change_prob(fund_quarters, strategy="Private Equity")
vc_pchg = build_yearly_change_prob(fund_quarters, strategy="Venture Capital")
all_pchg = build_yearly_change_prob(fund_quarters, strategy=None)

# Load 1y transition matrices derived from 10y (preferred)
p1_all_path = os.path.join(DATA_DIR, "grade_transition_1y_all.csv")
p1_pe_path = os.path.join(DATA_DIR, "grade_transition_1y_pe.csv")
p1_vc_path = os.path.join(DATA_DIR, "grade_transition_1y_vc.csv")
P1_ALL = pd.read_csv(p1_all_path, index_col=0) if os.path.exists(p1_all_path) else None
P1_PE = pd.read_csv(p1_pe_path, index_col=0) if os.path.exists(p1_pe_path) else None
P1_VC = pd.read_csv(p1_vc_path, index_col=0) if os.path.exists(p1_vc_path) else None

def get_transition_matrix(strategy):
    # Prefer 1y matrices derived from 10y root if available
    if strategy == "Private Equity" and P1_PE is not None:
        return P1_PE, "PE_1Y"
    if strategy == "Venture Capital" and P1_VC is not None:
        return P1_VC, "VC_1Y"
    if P1_ALL is not None:
        return P1_ALL, "ALL_1Y"
    # Fallback to quarterly-derived matrix
    if strategy == "Private Equity" and pe_n >= MIN_TRANSITIONS:
        return pe_probs, "PE"
    if strategy == "Venture Capital" and vc_n >= MIN_TRANSITIONS:
        return vc_probs, "VC"
    return all_probs, "ALL"

def get_yearly_change_prob(strategy):
    if strategy == "Private Equity" and pe_n >= MIN_TRANSITIONS:
        return pe_pchg
    if strategy == "Venture Capital" and vc_n >= MIN_TRANSITIONS:
        return vc_pchg
    return all_pchg

def sample_next_grade(curr_grade, P_df, rng):
    if curr_grade not in GRADE_STATES:
        # default if missing/invalid
        return "D"
    row = P_df.loc[curr_grade].values
    return rng.choice(GRADE_STATES, p=row)

# Projection loop (ROBUST to NaNs)
# -----------------------------
rng = np.random.default_rng(1234)

out = []
for _, r in base_rows.iterrows():
    fund_id = r["FundID"]
    nav = float(r["NAV_start"])  # always numeric now
    age0 = int(r["Fund_Age_Quarters"]) if pd.notna(r["Fund_Age_Quarters"]) else 0

    strategy = r["Adj Strategy"]
    grade = r["Grade"]
    cap_qe = r["cap_qe"]
    if pd.isna(cap_qe):
        cap_qe = msci_future["quarter_end"].iloc[-1]

    for step, (qe, msci_r, msci_r_lag1) in enumerate(
        zip(msci_future["quarter_end"], msci_future["msci_ret_q"], msci_future["msci_ret_q_lag1"]),
        start=1
    ):
        if qe > cap_qe:
            break

        # harden lag
        msci_r_lag1 = 0.0 if pd.isna(msci_r_lag1) else float(msci_r_lag1)

        age = age0 + step
        age_bucket = make_age_bucket_q(pd.Series([age])).iloc[0]

        # --- Update grade yearly using transition matrix + change probability ---
        prev_grade = grade
        if step % 4 == 0:
            pchg = get_yearly_change_prob(strategy)
            if rng.random() < pchg:
                P, src = get_transition_matrix(strategy)
                grade = sample_next_grade(grade, P, rng)

        b0, b1, _ = get_betas(strategy, grade)
        alpha, _ = get_alpha(strategy, grade, age_bucket)
        sigma, _ = get_sigma(strategy, grade)

        eps = rng.standard_normal()
        omega = alpha + b0 * float(msci_r) + b1 * msci_r_lag1 + sigma * eps

        # final hardening
        if not np.isfinite(omega):
            omega = 0.0
        omega = float(np.clip(omega, -OMEGA_CLIP, OMEGA_CLIP))

        nav = max(nav * (1.0 + omega), 0.0)

        out.append({
            "FundID": fund_id,
            "quarter_end": qe,
            "msci_ret_q": float(msci_r),
            "msci_ret_q_lag1": float(msci_r_lag1),
            "omega": omega,
            "NAV_projected": float(nav),
            "Fund_Age_Quarters": int(age),
            "Adj Strategy": strategy,
            "Grade_prev": prev_grade,
            "Grade": grade,
            "AgeBucket": age_bucket,
            "cap_qe": cap_qe,
        })

nav_proj = pd.DataFrame(out)
print("\nProjected rows:", len(nav_proj))
print("Projected funds:", nav_proj["FundID"].nunique())

# Save simulated grade paths (cap at fund end + overrun)
grade_path = nav_proj.loc[nav_proj["quarter_end"] <= nav_proj["cap_qe"], ["FundID", "quarter_end", "cap_qe", "Adj Strategy", "Grade_prev", "Grade"]].copy()
grade_path_out = os.path.join(DATA_DIR, f"grade_path_simulated_{year}_{quarter}_{n_q}q.csv")
grade_path.to_csv(grade_path_out, index=False)
print("Saved grade paths:", grade_path_out)

print(nav_proj.head())
```
